```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.style.available
```

```python
df = pd.read_csv('results/result_0.csv')
fig, ax = plt.subplots(1, 1, figsize=(5, 4))

for half_mode in df['half_mode'].unique():
    df[df.half_mode == half_mode].plot(x='batch_size', y='frames_per_second', ax=ax, label=half_mode)
    ax.set_title('2d AE on A6000')
    ax.set_xlabel('batch size')
    ax.set_ylabel('frames per second')
    ax.set_ylim(1000, 7500)
fig.savefig('results/ae2d_inference_speed_A6000.png', bbox_inches='tight', transparent=False, dpi=300)

for mode in df.half_mode.unique():
    df_sub = df[ df.half_mode == mode ].drop(['gpu_id', 'half_mode'], axis = 1)
    df_sub.to_csv(f'results/inference_speed_A6000_2d_{mode}.csv', index = False)
```
